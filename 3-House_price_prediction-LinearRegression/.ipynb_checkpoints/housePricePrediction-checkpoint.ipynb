```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
houseData = sklearn.datasets.load_boston()
```

```python
data = pd.DataFrame(houseData.data, columns = houseData.feature_names)
data["price"] = houseData.target
```

```python
data.head()
```

```python
#checking null values
data.isnull().sum()
```

```python
correlation = data.corr()
```

```python
correlation
```

```python
X = data.drop(["price"], axis=1)
y = data["price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
predict = model.predict(X_test)
```

```python
#R squared error
score1 = metrics.r2_score(y_test, predict)

#mean absolute error
score2 = metrics.mean_absolute_error(y_test, predict)
```

```python
print(score1, score2)
```

```python
#plot
plt.scatter(y_test, predict)
plt.xlabel("actual price")
plt.ylabel("predicted price")
plt.show()
```
